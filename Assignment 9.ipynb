```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
file_path = r"D:\Data Science\Titanic.csv"

if os.path.exists(file_path):
    titanic_data = pd.read_csv(file_path)
    print(titanic_data.columns)
    print(titanic_data.describe())
    sns.countplot(x='Survived', data=titanic_data)
    plt.title('Survival Count')
    plt.show()
```

```python

print(titanic_data.columns)
```

```python
sns.countplot(x='Sex', hue='Survived', data=titanic_data)
plt.title('Survival Count by Gender')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=titanic_data)
plt.title('Survival Count by Passenger Class')
plt.show()
```

```python
plt.hist(titanic_data['Age'].dropna(), bins=30, color='blue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.scatter(titanic_data['Age'], titanic_data['Fare'], alpha=0.5)
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
sns.kdeplot(titanic_data.loc[titanic_data['Survived'] == 1, 'Age'].dropna(), shade=True, label='Survived', color='green')
sns.kdeplot(titanic_data.loc[titanic_data['Survived'] == 0, 'Age'].dropna(), shade=True, label='Not Survived', color='red')
plt.title('Age Distribution for Survival vs Non-Survival')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
sns.barplot(x='Pclass', y='Fare', data=titanic_data)
plt.title('Fare by Passenger Class')
plt.show()
```
